```python
import os
import glob

# --- 1. SETUP PATHS ---
hpc_root = "/rds/general/user/ak8224/home/emoji_project/data/Emotion"

# Directory containing your preprocessed audio files
wavs_dir = f"{hpc_root}/Male-Emotion/Angry"

# The single master text file containing all transcripts for this speaker
transcript_file = f"{hpc_root}/Male-Emotion/0013.txt"

# Where to save the final metadata file
output_metadata = f"{hpc_root}/Male-Emotion/metadata-male-angry.csv"

# The Speaker ID you want to assign to this specific emotion/speaker mapping
SPEAKER_ID = 0 

# --- 2. BUILD THE TRANSCRIPT DICTIONARY ---
transcript_map = {}
print(f"📖 Reading master transcripts from: {transcript_file}")

with open(transcript_file, "r", encoding="utf-8") as f:
    for line in f:
        # ESD format is usually tab-separated: filename \t transcript \t emotion
        parts = line.strip().split('\t')
        
        # Ensure the line has at least the filename and the text
        if len(parts) >= 2:
            file_id = parts[0].strip()
            text = parts[1].strip()
            
            # Save it to our lookup table
            transcript_map[file_id] = text

print(f"🧠 Memorized {len(transcript_map)} transcripts.")

# --- 3. GATHER AND MATCH AUDIO FILES ---
wav_files = glob.glob(os.path.join(wavs_dir, "*.wav"))
print(f"🔍 Found {len(wav_files)} audio files in the target directory.")

metadata_lines = []
missing_scripts = 0

for wav_path in wav_files:
    # Extract just the filename without the .wav extension (e.g., "0015_000001")
    base_name = os.path.splitext(os.path.basename(wav_path))[0]
    
    # Instantly look up the transcript in our dictionary
    if base_name in transcript_map:
        transcript = transcript_map[base_name]
        
        # Format: <absolute_wav_path>|<speaker_id>|<transcript>
        line = f"{wav_path}|{SPEAKER_ID}|{transcript}"
        metadata_lines.append(line)
    else:
        print(f"⚠️ Warning: Transcript missing for {base_name}.wav")
        missing_scripts += 1

# --- 4. SAVE TO CSV ---
with open(output_metadata, "w", encoding="utf-8") as f:
    for line in metadata_lines:
        f.write(line + "\n")

# --- 5. REPORTING ---
print(f"\n✅ Successfully generated metadata.csv with {len(metadata_lines)} entries!")
if missing_scripts > 0:
    print(f"ℹ️ Skipped {missing_scripts} audio files because they weren't in the transcript file.")
print(f"📄 Saved to: {output_metadata}")
```

## Combine Metadata

```python
def combine_metadata(file_paths, output_path, seed=6033689):
    """
    Combines multiple metadata files, assigns strict Speaker IDs, and perfectly shuffles the output.
    """
    all_lines = []
    
    # Enumerate gives us an automatic 0, 1, 2 counter for our Speaker IDs
    for speaker_id, path in enumerate(file_paths):
        print(f"🛠️ Processing {path} -> Assigning Speaker ID: {speaker_id}")
        
        with open(path, 'r', encoding='utf-8') as in_f:
            for line in in_f:
                line = line.strip()
                if not line:
                    continue
                
                parts = line.split('|')
                
                # Safety check to ensure it's a valid Matcha-TTS format row
                if len(parts) >= 3:
                    # Force the ID to match our new multi-speaker architecture
                    parts[1] = str(speaker_id)
                    new_line = '|'.join(parts)
                    all_lines.append(new_line)
                else:
                    print(f"⚠️ Skipping malformed line: {line}")
                    
    # 🔀 The Magic Step: Shuffle the entire combined dataset!
    print("🔀 Shuffling the combined dataset...")
    random.seed(seed) # Locks the randomness so you get the exact same shuffle if you re-run it
    random.shuffle(all_lines)
    
    # Write the beautifully mixed data to the final file
    with open(output_path, 'w', encoding='utf-8') as out_f:
        for line in all_lines:
            out_f.write(line + '\n')
            
    print(f"\n✅ Master metadata created at: {output_path} with {len(all_lines)} total rows.")
```

## Calculating Mel Statistics

```python
def get_combined_mel_stats(wav_dirs):
    print(f"🔍 Scanning {len(wav_dirs)} directories...")
    
    all_wav_files = []
    
    # 1. Loop through the list of directories and gather all file paths
    for wav_dir in wav_dirs:
        files = glob.glob(os.path.join(wav_dir, "*.wav"))
        all_wav_files.extend(files)
        print(f"  - Found {len(files)} files in {wav_dir}")
        
    if not all_wav_files:
        print("❌ No .wav files found! Check your directory paths.")
        return

    print(f"\n📊 Found {len(all_wav_files)} TOTAL files. Calculating Log-Mel Spectrograms...")

    # Match these EXACTLY to your yaml data parameters
    mel_transform = torchaudio.transforms.MelSpectrogram(
        sample_rate=22050,
        n_fft=1024,
        win_length=1024,
        hop_length=256,
        f_min=0.0,
        f_max=8000.0,
        n_mels=80,
        power=1.0, 
        normalized=False
    )

    all_mels = []
    
    for f in all_wav_files:
        waveform, sr = torchaudio.load(f)
        
        # Resample just in case any files aren't 22050Hz
        if sr != 22050:
            waveform = torchaudio.functional.resample(waveform, sr, 22050)
        
        # Convert to Mel
        mel = mel_transform(waveform)
        
        # Convert to Log-Mel 
        log_mel = torch.log(torch.clamp(mel, min=1e-5))
        all_mels.append(log_mel)

    print("\n🧠 Concatenating tensors...")
    # Stitch all audio frames together into one massive tensor
    all_mels_tensor = torch.cat(all_mels, dim=-1)
    
    # Calculate the global statistics
    mel_mean = all_mels_tensor.mean().item()
    mel_std = all_mels_tensor.std().item()

    print("\n✅ Calculation Complete!")
    print("-" * 40)
    print("data_statistics:")
    print(f"  mel_mean: {mel_mean:.6f}")
    print(f"  mel_std: {mel_std:.6f}")
    print("-" * 40)
```

## Splitting Validation

```python
def create_train_val_split(input_path, train_path, val_path, val_samples_per_speaker=5):
    val_lines = []
    train_lines = []
    speaker_val_counts = {}

    print(f"📖 Reading master metadata from: {input_path}...")
    
    with open(input_path, 'r', encoding='utf-8') as f:
        lines = [line.strip() for line in f if line.strip()]

    # The one-pass sorting loop
    for line in lines:
        parts = line.split('|')
        
        # Safety check to ensure it's a valid row
        if len(parts) >= 3:
            speaker_id = parts[1]
            
            # Initialize the counter if we haven't seen this speaker yet
            if speaker_id not in speaker_val_counts:
                speaker_val_counts[speaker_id] = 0
            
            # If this speaker still needs validation samples, route to val_lines
            if speaker_val_counts[speaker_id] < val_samples_per_speaker:
                val_lines.append(line)
                speaker_val_counts[speaker_id] += 1
            # Otherwise, route to train_lines (automatically deleting it from val)
            else:
                train_lines.append(line)
        else:
            print(f"⚠️ Skipping malformed line: {line}")

    # 💾 Save the beautifully balanced validation file
    with open(val_path, 'w', encoding='utf-8') as f:
        for line in val_lines:
            f.write(line + '\n')

    # 💾 Save the clean training file
    with open(train_path, 'w', encoding='utf-8') as f:
        for line in train_lines:
            f.write(line + '\n')

    # --- Print the Audit Summary ---
    print(f"\n✅ Split complete!")
    print(f"📂 Validation file saved: {val_path} ({len(val_lines)} lines)")
    print(f"📂 Training file saved: {train_path} ({len(train_lines)} lines)")
    
    print("\n📊 Validation Audit (Target: 5 per speaker):")
    # Sort the dictionary by integer Speaker ID for a clean readout
    for spk, count in sorted(speaker_val_counts.items(), key=lambda x: int(x[0])):
        print(f"   Speaker ID {spk}: {count} samples")
```

```python
# --- Execute the Split ---
input_master = "/rds/general/user/ak8224/home/MedEmoji-TTS/vector/emotion_master.txt"
train_output = "/rds/general/user/ak8224/home/MedEmoji-TTS/vector/emotion_master_train.txt"
val_output   = "/rds/general/user/ak8224/home/MedEmoji-TTS/vector/emotion_master_val.txt"

create_train_val_split(input_master, train_output, val_output)
```

## Handling Unicode Error

```python
def scrub_punctuation(file_path):
    print(f"🧹 Scrubbing {file_path}...")
    with open(file_path, 'r', encoding='utf-8') as f:
        content = f.read()

    # Replace full-width/smart characters with standard ASCII
    replacements = {
        '，': ',', '。': '.', '！': '!', '？': '?',
        '“': '"', '”': '"', '‘': "'", '’': "'"
    }
    
    for bad_char, good_char in replacements.items():
        content = content.replace(bad_char, good_char)

    with open(file_path, 'w', encoding='utf-8') as f:
        f.write(content)
    print("✅ Cleaned!")
```

## Pre-Process Audio

```python
def resample_dataset(metadata_paths, target_sr=22050):
    print(f"🎧 Starting batch resampling to {target_sr} Hz...")
    
    # 1. Collect all unique audio paths from your metadata files
    all_audio_paths = set()
    for path in metadata_paths:
        with open(path, 'r', encoding='utf-8') as f:
            for line in f:
                if line.strip():
                    audio_path = line.split('|')[0]
                    all_audio_paths.add(audio_path)
                    
    print(f"🔍 Found {len(all_audio_paths)} unique audio files to check.")
    
    # 2. Check and resample each file
    resample_count = 0
    for file_path in tqdm(all_audio_paths, desc="Processing Audio"):
        if not os.path.exists(file_path):
            print(f"⚠️ Missing file: {file_path}")
            continue
            
        # Load the audio
        waveform, current_sr = torchaudio.load(file_path)
        
        # If it doesn't match our target, convert and overwrite it!
        if current_sr != target_sr:
            resampler = torchaudio.transforms.Resample(orig_freq=current_sr, new_freq=target_sr)
            resampled_waveform = resampler(waveform)
            
            # Save it back out, overwriting the original file
            torchaudio.save(file_path, resampled_waveform, target_sr)
            resample_count += 1

    print(f"✅ Done! Successfully resampled {resample_count} files to {target_sr} Hz.")
```

```python
metadata_files = [
    "/rds/general/user/ak8224/home/MedEmoji-TTS/vector/emotion_master_train.txt",
    "/rds/general/user/ak8224/home/MedEmoji-TTS/vector/emotion_master_val.txt"
]

resample_dataset(metadata_files)
```
